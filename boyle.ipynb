```python
# Create a matrix with symptoms as rows and medications as columns. 
# Melt the dataframe to bring both Med1 and Med2 into a single column
melted_df = med_df.melt(id_vars='Symptom', value_vars=['Med1', 'Med2'], value_name='Medication')

#  heatmap data- groupby and pivoting
heatmap_data = melted_df.groupby(['Symptom', 'Medication']).size().unstack(fill_value=0)

```

```python


plt.figure(figsize=(12, 8))
ax = sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt="d")

# Adding "Frequency" to the colorbar legend
cbar = ax.collections[0].colorbar
cbar.set_label("Frequency")

plt.show()
```

```python


# List of medication classification columns
class_columns = ['Class_Antidepression', 'Class_Anticonvulsant', 'Class_Estradiol',
                 'Class_Progestin', 'Class_NK3']

# List of symptom columns
symptom_columns = ['Symp_NightSweats', 'Symp_SleepProb', 'Symp_HotFlash',
                   'Symp_LossBreastFullness', 'Symp_MoodChange']

# Create an empty DataFrame for heatmap data
heatmap2_data = pd.DataFrame(0, index=class_columns, columns=symptom_columns)

# Populate the heatmap data
for class_col in class_columns:
    for symptom_col in symptom_columns:
        heatmap2_data.at[class_col,
                        symptom_col] = med_df[med_df[class_col] == 1][symptom_col].sum()
```

```python

plt.figure(figsize=(10, 6))
sns.heatmap(heatmap2_data, annot=True, cmap='Blues', fmt='g')
plt.title('Heatmap of Symptoms by Medication Classification')
plt.show()
```
